```python
import pandas as pd
import numpy as np 
```

```python
df_1 = np.load('embeddings_1.npy')
df_2 = np.load('embeddings_2.npy')
```

```python
# Convert to DataFrames
df_1 = pd.DataFrame(df_1)
df_2 = pd.DataFrame(df_2)
```

```python
# Load and combine the ICD code files
icd_code_1 = pd.read_csv('icd_codes_1.txt', header=None)
icd_code_2 = pd.read_csv('icd_codes_2.txt', header=None)
```

```python
# Set column name for consistency
icd_code_1.columns = ['labels']
icd_code_2.columns = ['labels']
```

```python
icd_code_1.isnull().sum()
icd_code_2.isnull().sum()
```

```python
# Combine the ICD codes along rows (axis=0)
icd_code_3 = pd.concat([icd_code_1, icd_code_2], axis=0).reset_index(drop=True)
```

```python
# Change indexing to start from 1
icd_code_3.index = range(0, len(icd_code_3))
```

```python
labels=icd_code_3['labels'].str.split(';')
```

```python
labels
```

```python
Vocab=[]
```

```python
for i in labels:
    length=len(i)
    for j in range(0,length):
        Vocab.append(i[j])
```

```python
Vocab=list(set(Vocab))
```

```python
label_data=[]
for i in range(0,len(labels)):
    label_data.append(labels[i])
```

```python
label_data
```

```python
import tensorflow as tf

# Define the vocabulary
vocab = Vocab

# Create the StringLookup layer without an unknown token
lookup = tf.keras.layers.StringLookup(vocabulary=vocab, output_mode='multi_hot')

# Sample input data
input_data = tf.ragged.constant(label_data)

# Encode the input data
output = lookup(input_data)

print(output)
```

```python
output_train=pd.DataFrame(output)
```
